## Univariate visualizations

Quick visualizations of example columns referenced in blog post.

```python
# describe and visualize a single column
def explore_column(df, col_name):
    """
    Print a sample, the data type, number of NAs, and range
    of a column with name `col_name` in the dataframe `df`
    """
    print(f'Sample: {df[col_name].sample(5).values}')
    print(f'Type: {df[col_name].dtype}')
    print(f'NAs: {df[col_name].isna().sum()}')
    print(f'Range: {df[col_name].min()} to {df[col_name].max()}')
    sns.set_theme()
    sns.displot(data=df, x=col_name)
```

```python
# example for assessments of climate change
explore_column(df, 'percep_frost')
```

```python
# example for agricultural climate change mitigation
explore_column(df, 'legum')
```

```python
# example for personal values and preferences
explore_column(df, 'innov_early')
```

```python
# example for income and satisfaction
explore_column(df, 'thresh_ldw')
```

```python
# example for social network
explore_column(df, 'net_name3_imp')
```

```python
# example for risk preferences
explore_column(df, 'lott_1')
```

```python
# example for survey information
explore_column(df, 'date')
```
